```python
import os

path = 'C:\\Users\\Prashanth M\\Desktop\\Untitled Folder\\Data set\\git\\pulse\\data\\aggregated\\transaction\\country\\india\\state\\'
agg_trans_st_dir = os.listdir(path)
agg_trans_st_dir
```

```python
import pandas as pd
import json

agg_trans_dict = {'State':[],
                  'Year':[],
                  'Quater':[],
                  'Transacion_type':[],
                  'Transacion_count':[],
                  'Transacion_amount':[]}

for i in agg_trans_st_dir:
  st_path = path+i+'/'
  st_year = os.listdir(st_path)
  for j in st_year:   
    st_year_path = st_path+j+'/'
    st_year_dir = os.listdir(st_year_path)
    for k in st_year_dir:
      json_path = st_year_path+k
      jsonData = open(json_path, 'r')
      Data = json.load(jsonData)
      for x in Data['data']['transactionData']:
                    Name = x['name']
                    count = x['paymentInstruments'][0]['count']
                    amount = x['paymentInstruments'][0]['amount']
                    agg_trans_dict['Transacion_type'].append(Name)
                    agg_trans_dict['Transacion_count'].append(count)
                    agg_trans_dict['Transacion_amount'].append(amount)
                    agg_trans_dict['State'].append(i)
                    agg_trans_dict['Year'].append(j)
                    agg_trans_dict['Quater'].append('Q'+k.strip('.json'))

agg_trans_df = pd.DataFrame(agg_trans_dict)
#agg_trans_df
```

### By Brand

```python
path = "C:\\Users\\Prashanth M\\Desktop\\Untitled Folder\\Data set\\git\\pulse\data\\aggregated\\user\\country\\india\\state\\"
state_dir =os.listdir(path)

user_dict = {'State': [], 'Year': [], 'Quater': [], 'Brand': [],
    'Brand_count': [], 'Brand_percentage': []}

for i in state_dir:
  state_path = path+i+'/'
  year_dir  = os.listdir(st_path)
  for j in year_dir:
    year_path = state_path+j+'/'
    file_dir = os.listdir(year_path)
    for k in file_dir:
      file_path = year_path+k
      jsonFile = open(file_path,'r')
      Data = json.load(jsonFile)
      try:
        for x in Data['data']['usersByDevice']:
          brand = x['brand']
          count = x['count']
          percentage = x['percentage']

          user_dict['Brand'].append(brand)
          user_dict['Brand_count'].append(count)
          user_dict['Brand_percentage'].append(percentage)
          user_dict['State'].append(i)
          user_dict['Year'].append(j)
          user_dict['Quater'].append('Q'+k.strip('.json'))
      except TypeError:
        pass
      
userByBrand_df = pd.DataFrame(user_dict)

userByBrand_df
```

```python
userByBrand_df.to_csv('AggUserByBrand.csv',index=False)
```

Map transaction by states

```python
path = "C:\\Users\Prashanth M\\Desktop\\Untitled Folder\\Data set\\git\\pulse\\data\\map\\transaction\\hover\\country\\india\\state\\"


state_dir =os.listdir(path)


map_trans_dict = {'State': [], 'Year': [], 'Quater': [], 'District': [],
    'Transaction_count': [], 'Transaction_amount': []}

for i in state_dir:
  state_path = path+i+'/'
  year_dir  = os.listdir(st_path)
  for j in year_dir:
    year_path = state_path+j+'/'
    file_dir = os.listdir(year_path)
    for k in file_dir:
      file_path = year_path+k
      jsonFile = open(file_path,'r')
      Data = json.load(jsonFile)
      try:
        for x in Data['data']["hoverDataList"]:
          district = x['name']
          trans_count = x['metric'][0]['count']
          trans_amount = x['metric'][0]['amount']
          map_trans_dict['District'].append(district)
          map_trans_dict['Transaction_count'].append(trans_count)
          map_trans_dict['Transaction_amount'].append(trans_amount)
          map_trans_dict['State'].append(i)
          map_trans_dict['Year'].append(j)
          map_trans_dict['Quater'].append('Q'+k.strip('.json'))
      except TypeError:
        pass
      
mapTransByDist_df = pd.DataFrame(map_trans_dict)

mapTransByDist_df
```

```python
mapTransByDist_df.to_csv('mapTransByDistrict.csv',index=False)
```

```python
path= "C:\\Users\\Prashanth M\\Desktop\\Untitled Folder\\Data set\\git\\pulse\\data\\map\\user\\hover\\country\\india\\state\\"
state_dir =os.listdir(path)

mapUserDictReg = {'State': [], 'Year': [], 'Quater': [], 'District': [],
    'Registered_user': [], 'App_opening': []}

for i in state_dir:
  state_path = path+i+'/'
  year_dir  = os.listdir(st_path)
  for j in year_dir:
    year_path = state_path+j+'/'
    file_dir = os.listdir(year_path)
    for k in file_dir:
      file_path = year_path+k
      jsonFile = open(file_path,'r')
      Data = json.load(jsonFile)
      try:
        for x in Data['data']["hoverData"]:
          district = x
          registeredUser =  Data['data']["hoverData"][x]["registeredUsers"]
          app_opening = Data['data']["hoverData"][x]["appOpens"]
          mapUserDictReg['District'].append(district)
          mapUserDictReg['Registered_user'].append(registeredUser)
          mapUserDictReg['App_opening'].append(app_opening)
          mapUserDictReg['State'].append(i)
          mapUserDictReg['Year'].append(j)
          mapUserDictReg['Quater'].append('Q'+k.strip('.json'))
      except TypeError:
        pass
      
mapUserByDistReg_df = pd.DataFrame(mapUserDictReg)

mapUserByDistReg_df
```
