# Function

```python
def formatting(data_frame):
     ''' Sleact the data frame to format
     state and transaction_type'''


     if "State" in data_frame.columns:
          data_frame["State"] = data_frame["State"].apply(lambda x: x.replace("-", " "))
          data_frame['State'] = [i.title() for i in data_frame['State']]
     elif "state" in data_frame.columns:
          data_frame["state"] = data_frame["state"].apply(lambda x: x.replace("-", " "))
          data_frame['state'] = [i.title() for i in data_frame['state']]  

          return data_frame
     
     if "District" in data_frame.columns:
               data_frame['District'] = [str(i).title() for i in data_frame['District']]
               return data_frame
     
     elif "district" in data_frame.columns:
          data_frame['district'] = [str(i).title() for i in data_frame['district']]
          return data_frame
     
     if "Transacion_type" in data_frame.columns:
          data_frame["Transacion_type"] = data_frame["Transacion_type"].apply(lambda x: x.replace("-", " "))
          data_frame['Transacion_type'] = [i.title() for i in data_frame['Transacion_type']]
     else:
          return data_frame
```

```python
def map_state_id(data_frame):
    # Mapping the state id for state_transaction
    # Renaming the state name wrt geojson 
    data_frame['State'] = data_frame['State'].str.replace('Andaman & Nicobar Islands','Andaman & Nicobar Island')
    data_frame['State'] = data_frame['State'].str.replace('Dadra & Nagar Haveli & Daman & Diu', 'Dadara & Nagar Havelli')
    data_frame['State'] = data_frame['State'].str.replace('Delhi' , 'NCT of Delhi') 
    data_frame['State'] = data_frame['State'].str.replace('Arunachal Pradesh' , 'Arunanchal Pradesh') 
    data_frame['State'] = data_frame['State'].str.replace('NCT of NCT of Delhi' , 'NCT of Delhi') 
    data_frame['State'] = data_frame['State'].str.replace('Ladakh' , 'Jammu & Kashmir')#

    # Mapping the state ids in main data frame
    data_frame['id'] = data_frame['State'].apply(lambda x: state_id_map[x])

    return data_frame
```
